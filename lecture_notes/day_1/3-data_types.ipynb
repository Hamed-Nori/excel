# Datenarten und Formate in Excel

Excel arbeitet mit unterschiedlichen Datentypen, um Informationen strukturiert und übersichtlich darzustellen. In diesem Abschnitt gehen wir auf die Grundlagen der Dateneingabe ein – sei es Text, Zahl oder Datum. Erfahren Sie, wie Sie durch das richtige Formatieren Ihrer Daten präzise Berechnungen durchführen und übersichtliche Berichte erstellen können.

| Eigenschaft                    | Text                                                                                                    | Zahlen                                                                                                                   | Datum                                                                                                               |
|--------------------------------|---------------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------------|
| **Definition**                 | Zeichenfolge, bestehend aus Buchstaben, Zahlen und Sonderzeichen                                         | Numerische Werte, die für mathematische Berechnungen verwendet werden                                                     | Datums- und Zeitangaben, meist in Formaten wie TT.MM.JJJJ oder MM/DD/YYYY                                             |
| **Speicherformat**             | Wird als Zeichenkette (String) gespeichert                                                               | Intern als Ganzzahl oder Fließkommazahl gespeichert, abhängig von der Formatierung                                        | Intern als fortlaufende Zahl gespeichert (Seriennummer), aber benutzerfreundlich formatiert                           |
| **Standardausrichtung**        | Linksbündig                                                                                            | Rechtsbündig                                                                                                             | Oft zentriert oder rechtsbündig – kann je nach Einstellung variieren                                                |
| **Formatierungsmöglichkeiten** | Anpassbar in Schriftart, -größe, Farbe, Stil und mehr – jedoch nicht für numerische Berechnungen geeignet | Verschiedene Formate wie Standard, Währung, Prozent, Dezimalstellen, wissenschaftliche Notation; Möglichkeit der Tausendertrennung | Unterschiedliche Datumsformate (z. B. TT.MM.JJJJ, MM/DD/YYYY, ISO) sowie Uhrzeitdarstellungen möglich                |
| **Verwendung**                 | Für Bezeichnungen, Namen, Beschreibungen und andere Inhalte, die keine Berechnungen erfordern             | Für mathematische Operationen, Vergleiche, statistische Analysen und finanzielle Berechnungen                             | Für Terminplanung, Zeitrechnungen, historische Daten und zeitbasierte Analysen                                       |
| **Beispiel**                   | "Produktname", "Kunden", "Beschreibung"                                                                 | 123, 45.67, -8, 1000                                                                                                     | 01.01.2020, 12/31/2021, 15.03.2025                                                                                   |

## Kostenplaner

Um die Bedeutung und die Funktionsweise der verschiedenen Datentypen zu verstehen, werden wir gemeinsam einen Kostenplaner erstellen.

### Schritt 1

Anlegen einer leeren Arbeitsmappe und als "Kostenplaner" speichern.

### Schritt 2: Titel der Arbeitsmappe

In der Zelle **A1** wird in den meisten Fällen der Titel bzw. die Überschrift der jeweiligen Arbeitsseite definiert, um später einen Überblick zu haben.

### Schritt 3 (Texte)

- **Spalte "Kosten" definieren:** Unter dieser Spalte werden Kostenursachen wie Miete, Auto, Versicherung etc. aufgelistet; als letztes schreiben wir **Gesamt**.
- **Monatsnamen eintragen:** In derselben Reihe, in der "Kosten" steht, werden die Monatsnamen eingetragen. Mit der Taste **Tab** kann man horizontal zur nächsten Zelle wechseln.
- **Letzte Spalte:** Neben dem Monat Dezember wird als letzte Spalte **Gesamt** eingetragen.

### Schritt 4 (Zahlen)

- **Eintragen der Kosten:** In jede Zelle, je nach Monat, werden die Kosten ohne Währung eingetragen.
- **Nicht ausfüllen:** Die Zellen unterhalb der Spalte **Gesamt** bleiben leer.

<img src="../../assest\day_1\kostenplanner-2.png" alt="startbildschirm Excel">

Wie auf dem Bild zu sehen ist, unterscheidet sich die Formatierung von Text und Zahl. Während der Text linksbündig angezeigt wird, sind Zahlen rechtsbündig ausgerichtet. Dies sorgt für eine bessere Übersicht, insbesondere bei Zahlen mit vielen Nachkommastellen. Falls man die Zahlen linksbündig anzeigen möchte, kann man die Zellen markieren und über den Reiter **Start** die Ausrichtung ändern.

### Schritt 5 (Datum)

Bisher sind unsere Monatsangaben lediglich als Text hinterlegt. Excel erkennt sie daher nicht als Datum.  
Unser Ziel ist es, dass in jeder Monatszelle folgendes steht: **Monat 2024**.

- **Datentyp ändern:** Zunächst ändern wir den Datentyp von Text zu Datum, indem wir auf eine Monatszelle klicken und anschließend den Reiter **Start** auswählen.
- **Dropdown-Menü nutzen:** Rechts von der Sektion "Ausrichtung" befinden sich einige Elemente. Dort findet sich ein Dropdown-Menü mit dem Titel **Benutzerdefiniert**. Dieses öffnen wir.
- **Datum auswählen:** Im Dropdown-Menü wählen wir "Datum", sodass Excel erkennt, dass es sich um einen Datentyp handelt.

Allerdings zeigt Excel das Datum standardmäßig als **Jan 2024** an. Wir möchten jedoch **Januar 2025** darstellen. Um dies umzusetzen, führen wir folgende Schritte aus:

- Auf das Dropdown-Menü klicken.
- Ganz unten auf **Weitere Zahlenformate** klicken.
- Anschließend den Reiter **Benutzerdefiniert** auswählen.
- Im Eingabefeld **Typ** folgenden Formatcode eingeben: **MMMM JJJJ**.
- Diesen Vorgang für alle Monatszellen wiederholen, indem man sie markiert und die Schritte erneut ausführt.

### Schirtt 6 summe bilden

Bisher haben wir die verschidenen Datentyp kennengelernt.
Um die summe Monatskosten zur bilden benutzen wir die Zellendeklaration, welches wir bereits kennengelernt haben.
Dazu schreiben wir in Gesamt Zelle je nach Kategorie eine **=** zeichen und notieren wir den Zell Koordinaten und gebe eine **+** zeichen ein.
Anschlißend klicken wir auf Enter
