# Datenarten und Formate in Excel

Excel arbeitet mit unterschiedlichen Datentypen, um Informationen strukturiert und übersichtlich darzustellen. In diesem Abschnitt gehen wir auf die Grundlagen der Dateneingabe ein – sei es Text, Zahl oder Datum. Erfahren Sie, wie Sie durch das richtige Formatieren Ihrer Daten präzise Berechnungen durchführen und übersichtliche Berichte erstellen können.

| Eigenschaft                | Text                                                                                                    | Zahlen                                                                                                                   | Datum                                                                                                               |
|----------------------------|---------------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------------|
| **Definition**             | Zeichenfolge, bestehend aus Buchstaben, Zahlen und Sonderzeichen                                         | Numerische Werte, die für mathematische Berechnungen verwendet werden                                                     | Datums- und Zeitangaben, meist in Formaten wie TT.MM.JJJJ oder MM/DD/YYYY                                             |
| **Speicherformat**         | Wird als Zeichenkette (String) gespeichert                                                               | Intern als Ganzzahl oder Fließkommazahl gespeichert, abhängig von der Formatierung                                        | Intern als fortlaufende Zahl gespeichert (Seriennummer), aber benutzerfreundlich formatiert                           |
| **Standardausrichtung**    | Linksbündig                                                                                            | Rechtsbündig                                                                                                             | Oft zentriert oder rechtsbündig – kann je nach Einstellung variieren                                                |
| **Formatierungsmöglichkeiten** | Anpassbar in Schriftart, -größe, Farbe, Stil und mehr – jedoch nicht für numerische Berechnungen geeignet | Verschiedene Formate wie Standard, Währung, Prozent, Dezimalstellen, wissenschaftliche Notation; Möglichkeit der Tausendertrennung | Unterschiedliche Datumsformate (z. B. TT.MM.JJJJ, MM/DD/YYYY, ISO) sowie Uhrzeitdarstellungen möglich                |
| **Verwendung**             | Für Bezeichnungen, Namen, Beschreibungen und andere Inhalte, die keine Berechnungen erfordern             | Für mathematische Operationen, Vergleiche, statistische Analysen und finanzielle Berechnungen                             | Für Terminplanung, Zeitrechnungen, historische Daten und zeitbasierte Analysen                                       |
| **Beispiel**               | "Produktname", "Kunden", "Beschreibung"                                                                 | 123, 45.67, -8, 1000                                                                                                     | 01.01.2020, 12/31/2021, 15.03.2025                                                                                   |

## Kostenplaner

Um die Bedeutung und die FUnktionsweise von verschidene Datentypen zu verstehen werden wir zusammen einen Kostenplanner bauen.

### Schirtt 1

Anlegung von ein leeres Arbeitsmappe und als "Kostenplanner" speichern

### Schritt 2 Title der Arbeitsmappe

In der Zelle **A1** wird in den meistenfällen der Title, Überschrift der jeweiligen Arbeitsseite definiert um später einen Übersicht darauf zu haben.

### Schirtt 3 (Texte)

- Definierung von Spalten Kosten. Unter dieser Spalten definieren Kostenursachen wie Miete, Auto, Versicherung und etc. als Letztes schreiben wir **Gesamt**
- Im selben reihe wo Kosten steht schreiben wir Monatsnamen. Mit Taste **Tab** kann man die Zelle Horizontal wechseln
- Als letzte Spalten schreiben wir neben Monat Dezember  **Gesamt**

### Schirtt 4 (Zahlen)

- Kosten für jede Zelle je nach Monat eintragen ohne Währung
- Die Zellen unterhalb Spalte **Gesamt** nicht ausfüllen

<img src="../../assest\day_1\kostenplanner-2.png" alt="startbildschirm Excel" >

Wie man auf dem Bild sieht unterscheidet sich Die Formatierung von Text und Zahl. Während die Text linksbündig angezeigt werden werden Zahlen Rechtsbündig angezeigt. Das soll dafür sorgen dass man besseren Übersicht hat. vorallem bei zahlen die viele Nachkommastellen haben. Falls man die Zahlen Linksbündig haben will kann man die Zellen markieren und über den Reiter **Start** die Ausrichtung ändern.



### Schritt 5 (Datum)

Bisher sind unsere Monate nichts weiteres als eiene Text. Excel ist es nicht klar dass es sich dabei um einen Datum handelt.
Unsere Ziel ist es in jeder Monate Zelle foglendes steht: **Monat 2024**.

- als erstes ändern wir den Datentyp von Text zur Datum, dabei klicken wir auf eine Monatszelle, anschlißend auf Reiter **Start**
- Rechts von Sektion Ausrichtung sind paar Elemente zu sehen
- dort befindet sich eine Dropdown-Element mit Title **Benutzer Definiert**, klicken wir darauf
- im Dropdown menü auf Datum klicken
-Nun weiss Excel dass es sich um einen Datentyp Datum ändern. 

Allerdings zeigt Excel datum wie folgt **Jan 2024**. Wir wollen aber **Januar 2025**, um das umzusetzen führen wir folgende Schritte aus:

- Auf Dropdown klicken
- ganz unten auf **weitere Zahlenformate** klicken
- Anschlißend auf Benutzerdefiniert klicken
- Im Eingabefeld Typ folgendes schreiben: **MMMM JJJJ**
- jetzt für alle Zellen wiederholen in dem man alle Monatszellen marikiert und die schritte erneut ausführt
