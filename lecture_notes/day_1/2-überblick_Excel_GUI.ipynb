# Überblick über die Oberfläche

## Startbildschirm

<img src="../../assest\day_1\excel-startbildschirm.png" alt="startbildschirm Excel" hegiht="300px" width="500px">


Je nachdem, welche Version von Excel man nutzt, kann es sein, dass sich der Startbildschirm ein wenig unterscheidet. In unserem Fall verwenden wir hier die aktuelle Version Microsoft 365.  
Nun beginnen wir mit den Elementen, die man auf der linken Seite sieht. Dort befindet sich die Side-Bar. Hier finden wir die **Startseite**, **Neu** und **Öffnen**.  
Außerdem befinden sich ganz unten noch einige weitere Navigationsmenüs wie **Konto**, **Feedback** und **Optionen**.

<hr>
<img src="../../assest\day_1\side-bar.png" alt="startbildschirm Excel" max-height="250px" width="500px">

Im Header haben wir die Möglichkeit, eine leere Seite zu erstellen, eine Arbeitsmappe auszuwählen oder einige Tutorials anzusehen.  
Darunter haben wir die Möglichkeit, zuletzt verwendete, angeheftete oder als „Gefällt mir“ markierte Arbeitsmappen zu öffnen, sofern wir zuvor Excel benutzt haben. Außerdem gibt es eine Suchfunktion.  
Die Startseite kann sich, wie bereits erwähnt, je nach Version unterscheiden. Bei manchen Versionen wird direkt eine leere Arbeitsmappe geöffnet, was bei älteren Versionen der Fall war.  
Über das Navigationsmenü **Neu** ändert sich der Inhalt des Hauptbereichs. Dort hat man die Möglichkeit, spezifische Vorlagen zu verwenden, wie zum Beispiel **Persönliches Monatsbudget**.

<img src="../../assest\day_1\neu-seite-excel.png" alt="startbildschirm Excel" height="300px" width="500px">

Wir starten aber zunächst mit einer leeren Seite und gehen die einzelnen Elemente, die zu sehen sind, durch, um zu schauen, welche Möglichkeiten uns zur Verfügung stehen.

## Leere Arbeitsmappe

<img src="../../assest\day_1\Excel-leere-seite.jpg" alt="startbildschirm Excel" height="300px" width="500px">

Bevor wir mit den eigentlichen Arbeiten beginnen, kläre ich auf, was wir in unserem Workspace sehen können. Wir beginnen ganz oben. Dort haben wir eine Leiste für Schnellfunktionen.  
Von links beginnend sehen wir eine Option für die automatische Speicherung. Hier können wir festlegen, wo unser Projekt gespeichert wird. Als Nächstes haben wir einen Button zum Speichern. Für diese Funktion gibt es auch Shortcuts wie **Command + S**.  
Wie in vielen anderen Programmen haben wir als Nächstes zwei kreisförmige Pfeile, die in entgegengesetzte Richtungen zeigen. Diese beiden Buttons ermöglichen es uns, einen Schritt zurück oder vorwärts zu gehen, um Änderungen rückgängig zu machen bzw. wiederherzustellen. Dafür gibt es auch Shortcuts wie **Command + Z** oder **Command + Y**.  
Über das Dreipunkte-Menü kann man weitere Schnellfunktionen aufrufen, wie zum Beispiel Drucken, etc.

<img src="../../assest\day_1\header-excel.jpg" alt="startbildschirm Excel">

Unter der Schnellfunktionsleiste befindet sich die wichtigste Bedienfläche in Excel, nämlich das Menüband. Dieses Menüband ist in allen Office-Programmen enthalten. Das heißt, wer bereits mit PowerPoint oder Word gearbeitet hat, sollte dieses Menüband kennen.  
Dort haben wir in verschiedenen Tabs unterschiedliche Funktionen von Excel. Es ist nicht notwendig, jetzt über die einzelnen Punkte zu sprechen, denn diese werden wir im Laufe des Kurses nacheinander kennenlernen, nutzen und verstehen. Wichtig zu wissen ist, dass wir über das Menüband alle benötigten Funktionen erreichen können.  
Zum Beispiel können wir über den Reiter **Start** allgemeine Dinge wie Schriftart, Farbe und Ausrichtung kontrollieren und formatieren. Über den Reiter **Einfügen** können wir Elemente wie Tabellen, Diagramme, Kommentare etc. hinzufügen.  
Ein ganz wichtiger Reiter in Excel ist **Formeln**.  
Falls das Menüband viel Platz einnimmt, kann man es ausblenden, indem man auf den **Menüreiter**, in dem man sich befindet, klickt. Das ist ratsam bei Nutzern, die über einen kleineren Bildschirm verfügen.


## Bearbeitungsleiste & Tabelle

### Bearbeitungsleiste

Unterhalb unseres Menübands befindet sich die Bearbeitungsleiste. Diese ist im Laufe des Kurses sehr wichtig, da wir dort unsere Formeln eingeben.

<img src="../../assest\day_1\bearbeitungsleiste.png" alt="startbildschirm Excel">

Hier haben wir verschiedene Bereiche. Ganz links befindet sich unsere Feld-Auswahl. Standardmäßig steht dort **A1**, da in der darunterliegenden Tabelle auch das Feld **Spalte: A, Reihe: 1** ausgewählt ist.

Rechts davon haben wir das Eingabefeld für die Formeln. Wie das funktioniert, besprechen wir, sobald wir die Formeln anwenden. Über den Reiter **Ansicht > Bearbeitungsleiste** können wir die Bearbeitungsleiste ein- und ausblenden.

### Tabelle

Unterhalb der Bearbeitungsleiste befindet sich der Tabellenbereich, in dem wir die eigentlichen Daten, Bilder, Schaubilder etc. eintragen. Dieser Bereich wird auch Arbeitsmappe genannt, wobei eine Arbeitsmappe aus vielen Tabellen bestehen kann. Jede einzelne Zelle innerhalb der Tabelle besitzt einen eindeutigen Namen, der sich aus den Buchstaben A-Z (für die Spalte) und einer Zahl größer als 1 (für die Zeile) zusammensetzt. Somit hat man die Möglichkeit, Zellen untereinander zu referenzieren und miteinander zu verknüpfen.

## Arbeitsmappe

Das Tabellenblatt, in das man Eingabedaten eintragen kann, ist ein Excel-Arbeitsblatt. Dies ist ein Teil der Excel-Arbeitsmappe. Beim Start eines neuen Projekts besteht unsere Excel-Arbeitsmappe standardmäßig aus einem Excel-Arbeitsblatt mit dem Namen "Tabelle 1". Man kann beliebig viele Arbeitsblätter hinzufügen, bis die Rechenressourcen unseres Computers ausgelastet sind.  
Ein neues Arbeitsblatt kann man hinzufügen, indem man im Statusbereich unterhalb der Tabelle auf den **+**-Button klickt. Standardmäßig erhält das neue Arbeitsblatt den Namen **Tabelle-n**, wobei **n** eine fortlaufende Zahl ist. Den Namen kann man ändern, indem man einen Doppelklick darauf macht oder mit Rechtsklick und „Name ändern“. Per Drag & Drop kann man die Arbeitsblätter verschieben.  
Man kann das Arbeitsblatt auch löschen, indem man mit der rechten Maustaste darauf klickt und „Löschen“ wählt.  
In der Praxis besteht eine Arbeitsmappe meist aus mehr als einem Arbeitsblatt.

## Dateien speichern und öffnen

Um Dateien zu speichern oder zu öffnen, kann man über den Reiter **Datei** zum Startbildschirm wechseln. Dort sieht man, dass die Sidebar-Elemente erweitert wurden. Man hat dort Optionen wie **Speichern**, **Speichern unter**, **Drucken** etc.  
Klickt man auf **Speichern unter**, öffnet sich ein Dashboard, in dem man entscheiden kann, wo die Arbeitsmappe gespeichert wird.  
Man hat die Möglichkeit, eine Arbeitsmappe in der Cloud, z.B. auf **OneDrive**, hochzuladen oder lokal auf dem Computer zu speichern. Klickt man auf die Option **Dieser PC**, öffnet sich der Explorer, in dem man einen Ordner auswählen, der Arbeitsmappe einen Namen geben und auf Speichern klicken kann. Die Arbeitsmappe wird mit dem Dateityp **.xlsx** erzeugt und gespeichert, was der Standard-Dateityp ist.  
Früher, in Versionen ab 2007, gab es auch einen anderen Dateityp namens **.xls**.  
Man kann den Dateityp auch ändern, aber wir beginnen zunächst mit dem Standard-Dateityp. Nach dem Speichern sollte sich in der Schnellfunktionsleiste der Name unserer Arbeitsmappe geändert haben.

Um eine Datei zu öffnen, muss man im Startbildschirm auf der linken Sidebar auf die Option **Öffnen** klicken. Anschließend öffnet sich der Explorer (bei Windows) bzw. der Finder (bei macOS). Dort kann man zur gewünschten Arbeitsmappe navigieren, darauf klicken und sie öffnen.
